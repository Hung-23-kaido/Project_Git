```python
a = list(int(i) for i in input().split())
grade = {'positives':[], 'negatives':[],'zeros':[]}
for i in a:
    if i >0:
        grade['positives'].append(i)
    elif i < 0:
        grade['negatives'].append(i)
    else:
        grade['zeros'].append(i)
print(grade)
```
