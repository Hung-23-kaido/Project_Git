```python
#W8A2
class Date:
    def __init__(self, day, month, year):
        self.day = day
        self.month = month
        self.year = year

    def is_leap_year(self):
        return self.year % 400 == 0 or (self.year % 4 == 0 and self.year % 100 != 0)

    def days_in_month(self):
        days = [0, 31, 28, 31, 30, 31, 30,
                   31, 31, 30, 31, 30, 31]
        if self.month == 2 and self.is_leap_year():
            return 29
        return days[self.month]

    def is_valid(self):
        if self.year < 1 or self.month < 1 or self.month > 12:
            return False
        if self.day < 1 or self.day > self.days_in_month():
            return False
        return True

    def next_day(self):
        self.day += 1
        if self.day > self.days_in_month():
            self.day = 1
            self.month += 1
            if self.month > 12:
                self.month = 1
                self.year += 1

    def __str__(self):
        return f"{self.day:02d}/{self.month:02d}/{self.year}"

try:
    s = input().strip()
    d, m, y = map(int, s.split('/'))

    date = Date(d, m, y)

    if not date.is_valid():
        print("INVALID")
    else:
        date.next_day()
        print(date)

except:
    print("INVALID")
```

```python
#W8A3
class Point:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    
    def check_ori(self):
        return self.x == 0 and self.y == 0
    
    def check_x(self):
        return self.x == 0

    def check_y(self):
        return self.y == 0
    
    def distance(self):
        return (self.x**2+self.y**2)**(1/2)

x,y = map(int,input().split())
point = Point(x,y)
if point.check_ori():
    print("Goc toa do")
else:
    if point.check_x():
        print("truc hoanh")
    elif point.check_y():
        print("truc tung")
    else: print("tu do")
print(f'{point.distance():.2f}')
```

```python
#W8A6
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        return self.w * self.h

    def perimeter(self):
        return 2 * (self.w + self.h)

    def scale(self, k):
        self.w *= k
        self.h *= k


w, h, k = map(float, input().split())

r = Rectangle(w, h)
r.scale(k)

print(f"{r.area():.2f} {r.perimeter():.2f}")
```

```python
#W8A7
class Fraction:
    def __init__(self, num, den):
        if den < 0:
            num = -num
            den = -den
        a = abs(num)
        b = den
        while b != 0:
            a, b = b, a % b
        g = a
        self.num = num // g
        self.den = den // g

    def add(self, other):
        return Fraction(self.num * other.den + other.num * self.den,
                        self.den * other.den)

    def sub(self, other):
        return Fraction(self.num * other.den - other.num * self.den,
                        self.den * other.den)

    def mul(self, other):
        return Fraction(self.num * other.num,
                        self.den * other.den)

    def div(self, other):
        return Fraction(self.num * other.den,
                        self.den * other.num)

    def __str__(self):
        return str(self.num) + "/" + str(self.den)


a, b, op, c, d = input().split()
a = int(a)
b = int(b)
c = int(c)
d = int(d)

f1 = Fraction(a, b)
f2 = Fraction(c, d)

if op == "+":
    res = f1.add(f2)
elif op == "-":
    res = f1.sub(f2)
elif op == "*":
    res = f1.mul(f2)
else:
    res = f1.div(f2)

print(res)
```

```python
#W8A8
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def deposit(self, x):
        self.balance += x

    def withdraw(self, x):
        if x <= self.balance:
            self.balance -= x

    def __str__(self):
        return self.owner + ": " + format(self.balance, ".2f")


owner, balance = input().split()
balance = float(balance)

acc = BankAccount(owner, balance)

n = int(input())
for _ in range(n):
    cmd, x = input().split()
    x = float(x)
    if cmd == "DEPOSIT":
        acc.deposit(x)
    else:
        acc.withdraw(x)

print(f"{acc.balance:.2f}")
```

```python
#W8A9
class Student:
    def __init__(self, name):
        self.name = name
        self.scores = {}

    def add_score(self, subject, score):
        self.scores[subject] = score

    def avg(self):
        total = 0
        for s in self.scores:
            total += self.scores[s]
        return total / len(self.scores)

    def rank(self):
        a = self.avg()
        if a >= 8:
            return "Excellent"
        if a >= 6.5:
            return "Good"
        if a >= 5:
            return "Average"
        return "Poor"


name = input()
n = int(input())

st = Student(name)
for _ in range(n):
    subject, score = input().split()
    st.add_score(subject, float(score))

a = st.avg()
print(st.name, f"{a:.2f}", st.rank())
```

```python
#W8A10
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year


class Library:
    def __init__(self):
        self.books = []

    def add(self, book):
        self.books.append(book)

    def count_by_author(self, name):
        cnt = 0
        for b in self.books:
            if b.author == name:
                cnt += 1
        return cnt

    def find_by_year(self, y):
        cnt = 0
        for b in self.books:
            if b.year == y:
                cnt += 1
        return cnt


lib = Library()
n = int(input())

for _ in range(n):
    line = input()
    if line.startswith("ADD"):
        _, rest = line.split(" ", 1)
        title, author, year = rest.split(";")
        lib.add(Book(title, author, int(year)))
    elif line.startswith("COUNTYEAR"):
        _, y = line.split()
        print(lib.find_by_year(int(y)))
    else:
        _, author = line.split(" ", 1)
        print(lib.count_by_author(author))
```

```python
#W8A12
class ComplexNumber:
    def __init__(self, realPart, imaginaryPart):
        self.realPart = realPart
        self.imaginaryPart = imaginaryPart

    def multiply(self, other):
        real = self.realPart * other.realPart - self.imaginaryPart * other.imaginaryPart
        imag = self.realPart * other.imaginaryPart + self.imaginaryPart * other.realPart
        return ComplexNumber(real, imag)


a, b = map(float, input().split())
c, d = map(float, input().split())

z1 = ComplexNumber(a, b)
z2 = ComplexNumber(c, d)

res = z1.multiply(z2)
print(res.realPart, res.imaginaryPart)
```

```python
# W8A13
import math

class Triangle:
    def __init__(self, edge1, edge2, edge3):
        self.edge1 = edge1
        self.edge2 = edge2
        self.edge3 = edge3

    def getArea(self):
        a = self.edge1
        b = self.edge2
        c = self.edge3
        if a + b <= c or a + c <= b or b + c <= a:
            return -1
        p = (a + b + c) / 2
        return math.sqrt(p * (p - a) * (p - b) * (p - c))


a, b, c = map(int, input().split())

t = Triangle(a, b, c)
area = t.getArea()

if area == -1:
    print("invalid")
else:
    print(f"{area:.2f}")
```
