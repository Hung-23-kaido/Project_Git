```python
#W5A3
def is_prime(a):
    if a <= 1:
        return False
    for i in range(2,round(a**(1/2))+1):
        if a%i==0:
            return False
    return True

a = int(input(""))
if is_prime(a):
    print("Yes")
else:
    print("No")
```

```python
#W5A4
def perfect(a):
    sum = 0
    for i in range(1,a//2+1):
        if a%i == 0:
            sum+=i
    return sum == i

a = int(input(""))
if perfect(a):
    print("Yes")
else:
    print("No") 
```

```python
#W5A5
def find_first_index(lst, k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i
    return -1

p = list(int(i) for i in input().split())
k = int(input(""))
print(find_first_index(p, k))
```

```python
#W5A6
def gt(a):
    if a == 1:
        return 1
    return a*gt(a-1)

print(gt(5))
```

```python
#W5A10
a,b = list(map(str,input().split()))
dic = {}
flag = True
for i in range(len(a)):
    if a[i] not in dic:
        dic[a[i]] = b[i]
    if i in dic:
        if b[i] != a[i]:
            flag = False
print(flag)
```
