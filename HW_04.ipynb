```python
#W4A8
def dx(n):
    t = ""
    while n != 0:
        t += str(n%10)
        n //= 10
    return int(t)

def ktdx(n):
    return n == dx(n)

n = int(input())
while not ktdx(n):
    n += dx(n)
print(n)
```

```python
#W4A9
def check_set(n):
    t = ""
    o = n
    while n != 0:
        if str(n%10) not in t:
            t = str(n%10) + t
        n //= 10
    return int(t) == o

def check_cp(n):
    if n**(1/2)%1==0:
        return True
    return False

n = int(input())
d = 0
while n > 0:
    if check_cp(n) and check_set(n):
        print(n)
        d += 1
    n -= 1
if not d:
    print("no number")
```

```python
#W4A10
def Collatz(n,d):
    if n==1:
        return d
    if n %2 == 0:
        return Collatz(n//2,d+1)
    else: return Collatz(3*n+1,d+1)

n = int(input())
m = 0
while n != 0:
    m = max(m,Collatz(n,0))
    n -= 1
print(m)
```
