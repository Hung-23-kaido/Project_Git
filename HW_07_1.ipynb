```python
#W7A1
def bin_search(a, left, right, k):
    if left > right:
        return -1
    mid = (left + right) // 2
    if a[mid] > k:
        return bin_search(a, left, mid - 1, k)
    elif a[mid] < k:
        return bin_search(a, mid + 1, right, k)    
    else:
        return mid

a = list(map(int,input().split()))
k = int(input())
print(bin_search(a,0,len(a)-1,k))
```

```python
#W7A3
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

a = list(map(int,input().split()))
print("Bubble sort:",bubble_sort(a))
print("Selection sort:",selection_sort(a))
```

```python
#W7A4
def most_frequent_element(arr):
    if not arr:
        return None
    count_dict = {}
    for item in arr:
        if item in count_dict:
            count_dict[item] += 1
        else:
            count_dict[item] = 1
    max_count = 0
    result = None

    for item, count in count_dict.items():
        if count > max_count:
            max_count = count
            result = item
            
    return result

a = list(map(int,input().split()))
print(most_frequent_element(a))
```

```python
#W7A5
def count_pairs_optimized(arr, X):
    count = 0
    seen = {}
    for num in arr:
        target = X - num

        if target in seen:
            count += seen[target]

        if num in seen:
            seen[num] += 1
        else:
            seen[num] = 1
            
    return count

a = list(map(int,input().split()))
x = int(input())
print(count_pairs_optimized(a, x))
```

```python
#W7A10
def max_subarray_mod_no_import(nums, m):
    def find_position(arr, val):
        left, right = 0, len(arr)
        while left < right:
            mid = (left + right) // 2
            if arr[mid] <= val:
                left = mid + 1
            else:
                right = mid
        return left

    max_mod = 0
    current_prefix = 0
    sorted_prefixes = [0]

    for num in nums:
        current_prefix = (current_prefix + num) % m

        if current_prefix > max_mod:
            max_mod = current_prefix

        idx = find_position(sorted_prefixes, current_prefix)

        if idx < len(sorted_prefixes):
            p_old = sorted_prefixes[idx]
            result = (current_prefix - p_old + m) % m
            if result > max_mod:
                max_mod = result

        sorted_prefixes.insert(idx, current_prefix)
        print(sorted_prefixes)

    return max_mod

nums = list(map(int,input().split()))
m = int(input())
print(max_subarray_mod_no_import(nums, m))
```

```python
#W7A13
def longest_harmonious_subsequence(arr):
    freq = {}
    for x in arr:
        freq[x] = freq.get(x, 0) + 1

    res = 0
    for x in freq:
        if x + 1 in freq:
            res = max(res, freq[x] + freq[x + 1])

    return res


arr = list(map(int, input().split()))
print(longest_harmonious_subsequence(arr))
```

```python
#W7A15
def list_name(n):
    flag_before = True
    flag_after = False
    flag_last = False
    flag_first = False
    
    for i in range(n):
        name = str(input())
        if i == 0:
            first = name

        if flag_after:
            after = name
            flag_after = False

        if name == "Nemo":
            if i == 0:
                flag_before = False
                flag_last = True
            flag_after = True
            flag_before = False
            if i == n-1:
                flag_first = True
                flag_after = False

        if flag_before or (flag_last and (i == n-1)):
            before = name

    if flag_first:
        after = first
    return [before,after]

n = int(input())
name = list_name(n)
print(name[0] + " and " + name[1])
```
