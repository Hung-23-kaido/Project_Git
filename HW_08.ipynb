```python
#W7A21
def check(n,t):
    if t>12 or t<1:
        return False
    if t in [1,3,5,7,8,10,12]:
        if n<0 or n>31:
            return False
    elif t == 2:
        if n<0 or n>29:
            return False
    else:
        if n<0 or n>30:
            return False
    return True

n,t = list(map(int,input().split()))
day = [0,31,29,31,30,31,30,31,31,30,31,30,31]
if check(n,t):
    dayo = n
    for month in range(1,t):
        dayo += day[month]
    print(dayo)
else:   print("invalid")
```

```python
#W7A22
n = int(input())
list_name = []
for _ in range(n):
    name = str(input())
    list_name.append(name)
dem_ten = {}
for ho_ten in list_name:
    ten = ho_ten.strip().split()[-1] 
    if ten in dem_ten:
        dem_ten[ten] += 1
    else:
        dem_ten[ten] = 1

so_lan_max = 0
for so_lan in dem_ten.values():
    if so_lan > so_lan_max:
        so_lan_max = so_lan

ket_qua = []
for ten, so_lan in dem_ten.items():
    if so_lan == so_lan_max:
        ket_qua.append(ten)

ket_qua.sort()
print(ket_qua[0])
```

```python
#W7A25
def max_k(num_str,k):
    n = len(num_str)
    remove = n - k
    stack = []
    for digit in num_str:
        while stack and remove > 0 and stack[-1] < digit:
            stack.pop()
            remove -= 1
        stack.append(digit)
    return ''.join(stack[:k])


p = str(input(""))
k = int(input())
print(max_k(p,k))
```

```python
#W7A27
def ivb(s):
    stack = []
    match = {')': '(','}': '{',']': '['}
    for ch in s:
        if ch in match.values():
            stack.append(ch)
        elif ch in match:
            if not stack or stack[-1] != match[ch]:
                return False
            stack.pop()
    return len(stack) == 0

s = str(input())
if ivb(s):
    print("YES")
else: print("NO")
```

```python
#W7A28
def sang_nt(n):
    listnt = [1]*(n+1)
    listnt[0] = listnt[1] = 0
    for i in range(2,round((n)**(1/2))+1):
        if listnt[i] == 1:
            for j in range(i**2,n+1,i):
                listnt[j] = 0
    return listnt

n = int(input())
listnt = sang_nt(n)
listnt[n] = 1
inh = []
for i in range(len(listnt)):
    if listnt[i]:
        inh.append(i)
ma = 0
x = 0
for i in range(1,len(inh)):
    hi = inh[i]-inh[i-1]
    if hi>ma:
        ma = hi
        x = inh[i-1]
        y = inh[i]
print(" ".join(list(str(i) for i in range(x+1,y))))
```

```python
#W7A29
arr = list(map(int, input().split()))

max_len = 1
cur_len = 1
end = 0

for i in range(1, len(arr)):
    if arr[i] > arr[i - 1]:
        cur_len += 1
    else:
        if cur_len > max_len:
            max_len = cur_len
            end = i - 1
        cur_len = 1

if cur_len > max_len:
    max_len = cur_len
    end = len(arr) - 1

start = end - max_len + 1
print(*arr[start:end + 1])
```

```python
#W7A30
arr = list(map(int, input().split()))

dp = [1] * len(arr)
for i in range(len(arr)):
    for j in range(i):
        if arr[i] > arr[j] and dp[j] + 1 > dp[i]:
            dp[i] = dp[j] + 1

print(max(dp))
```

```python
#W7A31
s = list(input())
n = len(s)
i = 0
while i < n:
    j = i + 1
    pos = i
    while j < n:
        if s[j] < s[pos]:
            pos = j
        j += 1
    if pos != i:
        s[i], s[pos] = s[pos], s[i]
        break
    i += 1
print("".join(s))
```
